```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Electric Vehicle Sales by State in India.csv')
df.head(10)
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
 # Check for missing values
print(df.isnull().sum())
```

```python
# Total numbers of ev vehicles sale year by year
df['Year'].value_counts()
```

```python
#Total number of ev vehicles sale state wise
df['State'].value_counts()
```

```python
# Total vehicle classes sales
df['Vehicle_Class'].value_counts()
```

```python
# Total number of Vehicles Type
df['Vehicle_Type'].value_counts()
```

```python
# Total number of Vehicle Category
df['Vehicle_Category'].value_counts()
```

```python
df.drop(columns=['Year','Date']).describe()
```

```python
df.info()
```

```python
#Data visualisation

plt.figure(figsize=(6,4))
plt.title('Yearly Analysis of EV Sales in India')
sns.lineplot(x='Year', y='EV_Sales_Quantity', data=df,marker='o', color='b')
plt.xlabel('Year')
plt.ylabel('EV Sales')
```

```python
#Monthly Analysis of EV Sales in India
plt.figure(figsize=(6,4))
plt.title('Monthly Analysis of EV Sales in India')
sns.lineplot(x='Month_Name', y='EV_Sales_Quantity', data=df,marker='o', color='r')
plt.xlabel('Month')
plt.ylabel('EV Sales')
```

```python
#State wise EV Sales
plt.figure(figsize=(6,7))
plt.title('State-Wise Analysis of EV Sales')
sns.barplot(y='State', x='EV_Sales_Quantity', data=df,hue='State', palette='bright')
plt.xlabel('States')
plt.ylabel('EV Sales')
```

```python
# Vehicle class wise Sales
plt.figure(figsize=(15,4))
sns.barplot(x='Vehicle_Class', y='EV_Sales_Quantity',data=df,
hue='Vehicle_Class', palette='bright')
plt.title('Analysis by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('EV Sales')
plt.xticks(rotation=90)
```

```python
# Analysis of ev vehicle Category wise
plt.figure(figsize=(6,4))
sns.barplot(x='Vehicle_Category',
y='EV_Sales_Quantity',data=df, hue='Vehicle_Category',palette='bright')
plt.title('Analysis by Vehicle Category')
plt.xlabel('Vehicle Category')
plt.ylabel('EV Sales')
plt.xticks(rotation=0)
```

```python
# Analysis of ev vehicle type 

plt.figure(figsize=(6,4))
sns.barplot(x='Vehicle_Type', y='EV_Sales_Quantity',data=df,
hue='Vehicle_Type', palette='bright')
plt.title('Analysis by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('EV Sales')
plt.xticks(rotation=90)
```
